# Data

```python
import pandas as pd
import os
import glob

train_df = pd.read_csv("/kaggle/input/rsna-2024-lumbar-spine-degenerative-classification/train.csv")
train_df.head()
```

```python
root_dir = '/kaggle/input/rsna-2024-lumbar-spine-degenerative-classification/train_images'

file_paths = glob.glob(os.path.join(root_dir, '*', '*', '*.dcm'))

data = [(os.path.basename(os.path.dirname(os.path.dirname(filepath))),
         os.path.basename(os.path.dirname(filepath)),
         os.path.splitext(os.path.basename(filepath))[0])
        for filepath in file_paths]

images_df = pd.DataFrame(data, columns=['study_id', 'series_id', 'instance_number'])
images_df.head()
```

```python
columns_dtype_update = images_df.columns
images_df = images_df.astype({column:int for column in columns_dtype_update})
```

```python
train_images = pd.merge(images_df, train_df, how="inner", on="study_id")
train_images.head()
```

```python
train_series = pd.read_csv("/kaggle/input/rsna-2024-lumbar-spine-degenerative-classification/train_series_descriptions.csv")
train_series.head()
```

```python
train_series_images = pd.merge(train_images, train_series, how="inner", on=["study_id", "series_id"])
train_series_images.dropna(inplace=True)
train_series_images.head()
```

```python
train_subset = train_series_images[train_series_images["series_description"] == "Sagittal T1"]
train_subset = train_subset.drop(columns="series_description")
```

```python
subset_columns = filter(lambda x: x.startswith("spinal_canal_stenosis"), train_df.columns)
subset_columns = (["study_id", "series_id", "instance_number"] + list(subset_columns))
train_subset = train_subset[subset_columns]
train_subset = pd.get_dummies(train_subset, dtype=int)
train_subset.reset_index(drop=True, inplace=True)

train_subset.head()
```
